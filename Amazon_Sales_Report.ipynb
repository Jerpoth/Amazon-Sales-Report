## Customer Segmentation

```python
# customer Segmentation

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

customer=df.groupby('Order ID').agg({
    'Revenue':'sum',
    'Qty':'sum',
    'Date':'count'
}).rename(columns={'Order Date':'Order Frequency'})

scaler = StandardScaler()
customer_scaled = scaler.fit_transform(customer)

Kmeans = KMeans(n_clusters = 4, random_state=42)
customer['cluster'] = Kmeans.fit_predict(customer_scaled)

plt.figure(figsize=(12,6))
sns.scatterplot(x=customer['Revenue'],y=customer['Qty'],hue=customer['cluster'],palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Total Revenue')
plt.ylabel('Total Quantity')
plt.show()
```

* Cluster 0 (Purple): This cluster represents customers with relatively low total revenue and total quantity. These are likely low-value customers who                         don't spend much or buy many items.

* Cluster 1 (Teal): Similar to Cluster 0, these customers also have low total revenue and quantity, but potentially even lower than Cluster 0. They                         could be new customers or infrequent buyers.

* Cluster 2 (Green): There's only one customer in this cluster, which is an outlier. This customer had very low revenue.

* Cluster 3 (Yellow): This cluster shows customers with low to moderate total revenue but varying quantities. Some customers in this group purchase a                         high quantity of items despite their revenue being on the lower side.

* The clusters seem to differentiate customers primarily based on their spending and purchase quantity, identifying groups with varying levels of value   and purchasing habits.

# Sales Prediction

```python
# Sales Prediction

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error

sales_data = df[['Year','Month','Qty','Amount','Revenue']]
x = sales_data[['Year','Month','Qty','Amount']]
y = sales_data['Revenue']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, random_state=42)

rf_model = RandomForestRegressor(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)

y_pred = rf_model.predict(x_test)


# Model Evalution
print("Mean Square Error :", mean_squared_error(y_test,y_pred))
print("R-squared Score :", r2_score(y_test,y_pred))
```

```python
# Feature Importance
imp = rf_model.feature_importances_
plt.figure(figsize=(8,6))
sns.barplot(x=x.columns, y=imp, palette = 'viridis')
plt.title('Feature Importance in Sales')
plt.show()
```

```python
# Preprocessing the Data

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Aggregate Revenue by Date
daily_revenue = df.groupby('Date')['Revenue'].sum().reset_index()

# Feature Engineering
daily_revenue['Day'] = daily_revenue['Date'].dt.day
daily_revenue['Month'] = daily_revenue['Date'].dt.month
daily_revenue['Year'] = daily_revenue['Date'].dt.year

#lag Feature
daily_revenue['Lag_1'] = daily_revenue['Revenue'].shift(1)
daily_revenue['Lag_7'] = daily_revenue['Revenue'].shift(7)
daily_revenue = daily_revenue.dropna()

# Train Test Split
X = daily_revenue[['Day','Month','Year','Lag_1','Lag_7']]
Y = daily_revenue['Revenue']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

#Model Training
model = RandomForestRegressor(random_state=42)
model.fit(X_train, Y_train)

#Prediction
Y_pred = model.predict(X_test)

# Evaluation
mae = mean_absolute_error(Y_test,Y_pred)
print(f"Mean Absolute Error : {mae}")
```

```python
# Visualization

plt.figure(figsize=(10,5))
plt.plot(daily_revenue['Date'][-len(Y_test):],Y_test,label="Actual Revenue", marker='o')
plt.plot(daily_revenue['Date'][-len(Y_test):],Y_pred,label="Predicted Revenue", marker='x')
plt.title('Revenue Prediction')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python

# Model Evaluation

mae = mean_absolute_error(Y_test,Y_pred)
print(f"Mean Absolute Error(MAE) : {mae}")
mse = mean_squared_error(Y_test, Y_pred)
print(f"Mean Squared Error(MSE) :{mse}")
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error(RMSE) :{rmse}")
r2 = r2_score(Y_test,Y_pred)
print(f"R2 Score(R2) : {r2}")
```
